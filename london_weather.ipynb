The following lineplot fails to capture any tangible relationship between cloud cover and max temperature.
So does the density plot below.

```python
# Plot max_temp against cloud_cover, without regression line
sns.lmplot(x = 'cloud_cover', y = 'max_temp', data = weather,
          fit_reg = False)
plt.title('Cloud Cover vs Maximum Temperature', loc='left')
plt.show()
```

```python
sns.kdeplot(data = weather, x="cloud_cover", y="max_temp", cmap="Reds", shade=True)
plt.title('Cloud Cover vs Maximum temperature', loc='left')
plt.show()
```

The following lineplot on the other hand does suggest that there is a negative correlation between cloud cover and
sunshine. The density plot confirms our intuition!

```python
# Lineplot of sunshine vs cloud cover
sns.lmplot(x = 'cloud_cover', y = 'sunshine', data = weather,
          fit_reg = False)
```

```python
# Density of sunshine vs cloud cover
sns.kdeplot(data = weather, x="cloud_cover", y="sunshine", cmap="Reds", shade=True)
plt.title('Precipitation vs mean temperature', loc='left')
plt.show()
```

```python
# Snow depth vs max temp
sns.lmplot(x = 'max_temp', y = 'snow_depth', data = weather,
          fit_reg = False)
plt.title('Snow depth vs Maximum Temperature', loc = 'left')
plt.show()
```

```python
sns.lmplot(x = 'mean_temp', y = 'precipitation', data = weather,
          fit_reg = False)
plt.title('Precipitation vs mean temperature', loc='left')
plt.show()
```

The above plot does little to reveal any link between mean temperature and precipitation.

```python
sns.kdeplot(data = weather, x="mean_temp", y="precipitation", cmap="Reds", shade=True)
#ax.set_ylim((0,4)
plt.ylim(0, 4)
plt.xlim(None, None)
 

plt.title('Precipitation vs mean temperature', loc='left')
plt.show()
```

This density plot however reveals that around on days with mean temperature of 5-20 degrees, there is frequently
upto 0.5 mm of rainfall. This is further confirmed by the following jointplot.

```python
sns.jointplot(data = weather, x="mean_temp", y="precipitation", kind="hex")

plt.ylim(0, 1)
plt.xlim(None, None)
plt.suptitle("Precipitation vs mean temperature")
plt.show()
```

The positive correlation between mean temperature and maximum temperature, as well as that between mean temperature 
and minimum temperature, are captured by next two stripplots.

```python
# Stripplot max_temp vs mean_temp
sns.stripplot(data =weather, x="mean_temp", y="max_temp", jitter=0.2, size=2)
plt.title('Max temperature vs mean temperature', loc='left')
plt.show()
```

```python
sns.stripplot(data =weather, x="mean_temp", y="min_temp", jitter=0.2, size=2)
plt.title('Min temperature vs Mean temperature', loc='left')
plt.show()
```

Further, the 'Min temperature vs mean temperature' stripplot is thinner than the 'Max temperature vs mean temperature', 
showing us greater variance in maximum temperature.

```python
sns.kdeplot(data = weather, y="sunshine", x="global_radiation", cmap="Reds", shade=True)
plt.title('Sunshine vs Global Radiation', loc='left')
plt.show()
```

This is just an attempt to capture the correlation between global radiation and sunshine.

```python
# Global radiation vs cloud cover
sns.kdeplot(data= weather, y="cloud_cover", x = "global_radiation", cmap = "OrRd")
plt.title('Cloud Cover vs Global Radiation', loc='center')
plt.show()
```

```python
# Boxplot of global_radiation
sns.boxplot(data = weather[["global_radiation"]], palette = ['#C03028'])
plt.title('Boxplot of Global Radiation')
plt.show()
```

What stands out in the above boxplot is the variation between the median and the upper quartile.

```python
sns.jointplot(x='mean_temp', y='pressure', data=weather)
plt.title('Pressure vs Mean Temperature', loc='left')
plt.show()
```

```python
sns.kdeplot(data = weather, x='mean_temp', y='pressure', cmap="Reds", shade=True)
plt.title('Pressure vs Mean Temperature', loc='left')
plt.show()
```

The density plot above shows the expected barometric pressure (in Pascals), especially between 15 to 20 degrees Celcius.
